```python
import cv2

cap = cv2.VideoCapture('video1.mp4')

while cap.isOpened():
    r, frame = cap.read()
    if r:
        fram = cv2.resize(frame, (600, 500))
        cv2.imshow('durgesh', fram)  # use resized frame

        # Small delay for video playback (40ms ~ 25 FPS)
        if cv2.waitKey(40) & 0xFF == ord("p"):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
